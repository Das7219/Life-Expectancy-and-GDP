## Introduction
The goal of this project is to compare life expectancy with GDP in individual countries and to address possible relationships between them.

```python
#Step 1 Import the modules that you'll be using in this project
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#Step 2 Load the dataset and inspect it
Life = pd.read_csv('all_data.csv')
print(Life.head(10))
print(Life.info())
```

```python
#Step 3 Compare GDPs of the countries over time, in order to eventually look for correlation between GDP and life expectancy
f, ax = plt.subplots(figsize=(15, 6)) 
ax = sns.pointplot(x="Year", y="GDP", hue = "Country", data=Life)
```

```python
# We can see at there is strongly correlation with the life expectancy and GDP. USA has high GDP and long life expectancy.
#Since 2006 in China we can observed increasing of life expectancy and GDP. In other countries we don't see huge changes.
```

```python
#Step 4 Compare GDP and life expectancy over time - plot the life expectancy data.
```

```python
f, ax = plt.subplots(figsize=(12, 6)) 
ax = sns.pointplot(x="Year", y="Life expectancy at birth (years)", hue = "Country", data=Life)
ax.set(ylabel="Life expectancy at birth (years)")
```

```python
#In this plot we see at developed countries have life expectency between 70-85 years. Zimbabwe which isn't develop country has life expectancy between 40-60 years.Since 2006, this life expectancy has increased significantly in Zimbabwe.
#This can be evidence of a positive development process in this country.
```

```python
#Step 5 Create bar charts for each DataFrame.
```

```python
ax = sns.barplot(x="Country", y="GDP", data=Life)
plt.xticks(rotation=90)
```

```python
ax = sns.barplot(x="Country", y="Life expectancy at birth (years)", data=Life)
plt.xticks(rotation=90)
plt.show()
```
